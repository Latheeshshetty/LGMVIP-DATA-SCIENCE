## IMAGE TO PENCIL SKETCH WITH PYTHON

## Importing libraries

```python
import cv2
import matplotlib.pyplot as pt
```

## loading the original image

##### first step is to load the image using imread() method and cvtColor() used to convert an image from one color space to another.
##### here BGR(blue,gray,red) to RGB(red,gray,blue)

```python
img=cv2.imread('C:\\Users\\dell\\Pictures\\abd.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
pt.figure(figsize=(8,8))
pt.axis("off")
pt.title("original image")
pt.imshow(img)
```

## converting the image to grayscale

##### here we need to convert original image to grayscale.
##### cmap in imshow() is a colormap instance or registered colormap name.

```python
im_gry=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
pt.figure(figsize=(8,8))
pt.axis("off")
pt.imshow(im_gry,cmap="gray")
pt.title("grayscale image")
pt.show()
```

## Inverting the image

##### for inverting the image we can use bitwise_not() function.

```python
im_invert=cv2.bitwise_not(im_gry)
pt.figure(figsize=(8,8))
pt.axis("off")
pt.imshow(im_invert,cmap="gray")
pt.title("inverted image")
pt.show()
```

## smoothing the image(inverted blur image)

##### GaussianBlur() method used to blur an image.

```python
im_smooth=cv2.GaussianBlur(im_invert, (21,21),sigmaX=0, sigmaY=0)
pt.figure(figsize=(8,8))
pt.axis("off")
pt.imshow(im_smooth,cmap="gray")
pt.title("Smoothen Image")
pt.show()
```

## converting image to pencil sketch

##### divide() function will divide the grayscale image and inverted blur image.

```python
im_sketch=cv2.divide(im_gry,255-im_smooth,scale=255)
pt.figure(figsize=(8,8))
pt.axis("off")
pt.imshow(im_sketch,cmap="gray")
pt.title("final sketch Image")
pt.show()
```
